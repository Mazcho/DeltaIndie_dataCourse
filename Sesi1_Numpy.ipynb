vstack, hstack, hsplit

```python
a_1 = np.array([[1,1],[2,2]])
a_1
```

```python
a_2 = np.array([[5,6],[7,8]])
a_2
```

```python
#vstack --> menggabungkan 2 array menjadi satu dengan susunan vertikal
result_vstack = np.vstack((a_1,a_2))
result_vstack
```

```python
result_vstack.shape
```

```python
#hstack digunakan untuk menggabungkan 2 array secara horizontal
#arr_1 = [[1,1],[2,2]]
#arr_2 = [[5,6],[7,8]]
#--> hstack output [ [1,1,2,2],
#                     [5,6,7,8]
#                    ]

result_hstack = np.hstack((a_1,a_2))
result_hstack
```

```python
a_3 = np.array([[5,10],[7,5]])
a_4 = np.array([[6,3],[8,4]])

vstack_2 = np.vstack((a_3,a_4))
hstack_2 = np.hstack((a_3,a_4))

print(vstack_2)
print(hstack_2)
```

Operator matematika numpy

```python
bilangan1 = np.array([1,2,3])
bilangan2 = np.array([10,5,4])

#pertambahan
result = bilangan1 - bilangan2
result
```

```python
#pengurangan
bilangan1 = np.array([8,2,3])
bilangan2 = np.array([1,2,3])
result_kurang = bilangan1 - bilangan2
result_kurang
```

```python
#perkalian
bilangan1 = np.array([8,2,3])
bilangan2 = np.array([1,2,3])
result_kali = bilangan1 * bilangan2
result_kali
```

```python
#pembagian
bilangan1 = np.array([8,2,3])
bilangan2 = np.array([1,2,3])
result_bagi = bilangan1 // bilangan2
result_bagi
```
